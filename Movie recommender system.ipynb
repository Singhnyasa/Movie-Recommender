```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('C:\\Users\\Nyasa\\OneDrive\\Desktop\\tmdb_5000_movies.csv')
credits = pd.read_csv('C:\\Users\\Nyasa\\OneDrive\\Desktop\\tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L  # Finding 1st 3 cast 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L   #finding director from crew
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new = movies[['movie_id','title','tags']]
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer
```

```python
def stem(text):
    y= []
    
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
